## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns

whale_returns_csv = Path("Resources/whale_returns.csv")
whale_returns = pd.read_csv(whale_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)

whale_returns
```

```python
whale_returns.columns
```

```python
# Count nulls

whale_returns.isnull().sum()
```

```python
# Drop nulls

whale_returns.dropna(inplace=True)
whale_returns.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
algo_returns = pd.read_csv(algo_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)

algo_returns
```

```python
# Count nulls

algo_returns.isnull().sum()
```

```python
# Drop nulls

algo_returns.dropna(inplace=True)
algo_returns.isnull().sum()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# combined = pd.concat([whale_returns, algo_returns], axis="columns", join="inner")
# combined.head()
```

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")
sp500_history = pd.read_csv(sp500_history_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)

# sp500_history.index = pd.to_datetime(sp500_history.index)

sp500_history.sort_index(inplace=True) # What does this do?

# sp500_history = sp500_returns.sort_index(ascending=True)

sp500_history
```

```python
# Check Data Types

sp500_history.dtypes
```

```python
# Fix Data Types

sp500_history['Close'] = sp500_history['Close'].replace(['\$'], [''], regex=True)
sp500_history['Close'] = sp500_history['Close'].astype('float64')

sp500_history
```

```python
sp500_history.plot(figsize=(30,10))
```

```python
sp500_history.dtypes
```

```python
# Calculate Daily Returns

sp500_returns = sp500_history.pct_change()
sp500_returns.head()
```

```python
# Drop nulls

sp500_returns.dropna(inplace=True)
sp500_returns.isnull().sum()
sp500_returns
```

```python
# Rename Column

columns = ["S&P 500"]
sp500_returns.columns = columns
sp500_returns.head()

# sp500_returns.columns
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame

combined1 = pd.concat([whale_returns, algo_returns, sp500_returns], axis="columns", join="inner")
combined1.head()
```
